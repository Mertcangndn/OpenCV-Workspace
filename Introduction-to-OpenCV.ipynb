## Bir Resmi Maskelemek
- Maskeleme bir resmin belli bir bölümünü izole etmeye yarar. Örneğin, bir cismi kırmızıyla sınırlandırmak gibi.
- Bu işlem için HSV kullanılır. HSV, RGB'nin aksine gölgelerle uğraşmadan sadece temel renkler üzerinden renkleri ayırabilmemizi sağlar:
    - **H**ue: Renk
    - **S**aturation: Yoğunluk
    - **V**alue: Yoğunluk

### Resmi HSV'ye Dönüştürme
- OpenCV tarafından okunan görseller default olarak BGR modundadır, bunu HSV'ye dönüştürmek için ```cvtColor()``` fonksiyonunu kullanacağız

````python
bird = cv2.imread('bird.png')
bird_hsv = cv2.cvtColor(       bird , cv2.COLOR_BGR2HSV)
#Türünü değiştireceğimiz görsel<-^          ^->Neyi neye dönüştüreceğimiz 
````

### Renk Aralığı Belirleme

````python
import numpy as np #Bu işlem için Numpy kütüphanesine ihtiyacımız var.

#Kırmızı renkli aralığı belirleme
#                     |-> OpenCV C++ ile yazılmıştır ve Python'daki karmaşık liste yapısını anlayamaz
#                     |-> bu yüzden numpy'deki array() fonksiyonu ile listeyi diziye çeviririz.
lower_red = np.array([0, 120, 70])
upper_red = np.array([10, 255, 255])
#   Bu aralık, hue değeri olarak 0'dan 10'a kadar,
#   saturation değeri olarak 120'den 255'ye kadar ve
#   value değeri olarak 70'den 255'ye kadar değer alır.
````

**Neden 0-10 Hue Değeri?**

Bu kodda, hue değeri olarak 0'dan 10'a kadar değeri alındığı için, kırmızı renkli aralık belirlenir. Çünkü, HSV renklendirme 
alanı içindeki hue değeri, renkin tonunu belirir. Kırmızı renk, hue değeri olarak 0-10 arasında yer alır.

**Neden 120-255 Saturation Değeri?**

Bu kodda, saturation değeri olarak 120'den 255'ye kadar değeri alındığı için, kırmızı renkli aralık belirlenir. Çünkü, HSV 
renklendirme alanı içindeki saturation değeri, renğin yoğunluğunu belirir. Kırmızı renk, yüksek yoğunlukta yer alır.

**Neden 70-255 Value Değeri?**

Bu kodda, value değeri olarak 70'den 255'ye kadar değeri alındığı için, kırmızı renkli aralık belirlenir. Çünkü, HSV 
renklendirme alanı içindeki value değeri, renğin şiddetiyle birlikte gösterir. Kırmızı renk, yüksek şiddetle yer alır.

### Mask Oluşturma ve Uygulama

```inRange()```:
- Resmin belli bir bölümünü izole etmek için kullanılan mask değişkenini oluşturmada kullanılır.
- If-else bloklarındaki gibi her pixel'i gezer ve istenilen aralıkta olanları beyaz, diğerlerini siyah yapar.
- Sonuç olarak ```mask``` değişkeni istenilen yerleri beyaz, diğer kısımlar siyah olan bir matris olur.
- Buna Binary Mask (İkili Maske) denir. Bu sadece bir şablondur.

```bitwise_and()```:
- ```mask``` matrisini resme uygular.
- Burada dikkat edersek _AND_ logic gate'i vardır.
- Bu fonksiyonun parametreleri şöyledir: (Kaynak1, Kaynak2, Maske).
1. ```image``` ve ```image```: Fonksiyon önce bu iki resmi birbirleriyle _AND_ işlemine sokar. Bir sayının kendisiyle AND işlemi yine kendisidir ```(A & A = A)```. Yani burası aslında formalitedir, resmin kendisini elde ederiz.
2. ```mask=mask```: OpenCV'ye şunu dersin: "Bu AND işlemini sadece maskenin BEYAZ (255) olduğu yerlerde **yap**, SİYAH (0) olduğu yerleri **direkt sıfırla**."

Matematiksel olarak şu olur:
- Maske Siyahsa (0): ```Renkli Piksel *AND* 0 = 0 (Siyah)``` _(Arka plan silindi, simsiyah oldu)_
- Maske Beyazsa (255 - Binary olarak hepsi 1): ```Renkli Piksel *AND* 11111111 = Renkli Piksel``` _(Rengin kendisi korundu, aynen kopyalandı)_
- Bu mantık ```bitwise_or()```, ```bitwise_not()``` ve ```bitwise_xor()``` fonksiyonları için de benzer şekilde çalışır.

````python
mask = cv2.inRange(bird_hsv, lower_red, upper_red)
result = cv2.bitwise_and(bird, bird, mask=mask) #BURADA HSV OLANI KULLANMADIĞIMIZA DİKKAT ET!!!

cv2.imshow("Red Bird",result)
cv2.imshow("Original Bird",bird)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

### Maskelenmiş Resimde Sınırları Oluşturma ```findContours()```
- Beyaz bölgelerin sınırlarını tek tek dolaşır ve bu sınırları oluşturan (x,y) noktalarını bir liste haline getirir.

```findContours()``` Parametreleri:
1. ```contours```: Bir "liste listesi"dir, Her eleman resimdeki ayrı bir beyaz adacığın sınır noktalarıdır.
2. ```_```: İlgilenmediğimiz veriler için python'da alt çizgi kullanırız.
3. ```RETR_EXTERNAL```: "Ben neyi alayım?" parametresidir. Sadece en dıştaki sınırları ver, detaylarla uğraştırma demektir.
    - ```RETR_EXTERNAL```: sadece simidin en dış yuvarlağını alır. Ortadaki deliği görmezden gelir.
    - ```RETR_TREE```: deseydin hem dışını hem de ortadaki deliği ayrı ayrı alırdı.
4. ```CHAIN_APPROX_SIMPLE```: Bellek optimizasyonu parametresidir.
    - ```CHAIN_APPROX_NONE``` kullansaydık tüm sınır pixellerinin koordinatlarını tek tek kaydederdi.
    - ```CHAIN_APPROX_SIMPLE``` ile sadece başlangıç ve bitiş pixellerinin (köşelerin) koordinatlarını kaydederiz.

Kısaca Kod Bize:_"Elimdeki maskeye bak. Beyaz adacıkların sadece dış sınırlarını bul. Bu sınırları bulurken de gereksiz ara noktaları at, bana sadece köşe noktalarını (sıkıştırılmış veriyi) getir ve konturlar listesine koy."_ der.

````python
colors = {
    "Blue": [
        np.array([0, 0, 150]), np.array([100, 100, 255])
    ],
    "Red": [
        np.array([180, 0, 0]), np.array([255, 80, 80])
    ],
    "Green": [
        np.array([0, 100, 0]), np.array([150, 255, 150]) 
    ],
    "Yellow": [
        np.array([200, 200, 0]), np.array([255, 255, 100])
    ],
    "Orange": [
        np.array([200, 100, 0]), np.array([255, 190, 100])
    ],
     "Light Blue": [
        np.array([0, 150, 150]), np.array([180, 255, 255])
    ] 
}
````

### Ağırlık Merkezi Hesaplama
- OpenCV ile sınırlarını aldığımız resimler sanki 3 boyutlu cisimlermiş gibi ağırlık merkezlerini hesaplayabiliriz.
- ```moments()``` fonksiyonu o denge noktasını hesaplamak için gereken verileri verir.
- Bu fonksiyon bir sözlük döndürür ve bu sözlüğün içinde bazı önemli bilgiler vardır:
    - ```M["m00"]``` (Alan/Kütle): Şeklin toplam alanı (piksel sayısı). Fizikteki "Kütle" gibi düşünebilirsin.
    - ```M["m10"]```: X eksenine göre moment (X koordinatlarının toplam ağırlığı).
    - ```M["m01"]```: Y eksenine göre moment (Y koordinatlarının toplam ağırlığı).

AĞırlık Merkezi Formülü:
- Cx​=M["m10"]/M["m00"] ​ ,  Cy​=M["m10"]/M["m00"]​
- Yani: (Toplam X(Y) Ağırlığı / Toplam Kütle) bize X(Y) merkezini verir.

````python
#--AZ ÖNCEKİ KISIM--
for color_name,range in colors.items():
    mask = cv2.inRange(image,range[0],range[1])
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    #--YENİ KISIM-- 
    for contour in contours:
        M = cv2.moments(contour)

        if M["m00"] != 0: #sıfıra bölünme hatası olmaması için kontrol
            cx = int(M["m10"]/M["m10"]) #Y ekseni koordinatları
            cy = int(M["m01"]/M["m00"]) #X ekseni koordinatları
        else:
            cx, cy = 0,0

        #Merkeze nokta koyma
        cv2.circle(outputImage, (cx,cy), 5, (0,0,0), -1) #cv2.circle(Tuval, Merkez, Yarıçap, Renk, Kalınlık)
        # -1 yerine 2 koyarsak kalınlığı 2 olan içi boş bir çember çizer, -1 olduğu için içini dolduruyor. 
````

### Maskelenmiş Görselin Etrafına Dikdörtgen Çizdirme ```boundingRect()```
- Elimizdeki kontur verisi sadece bir nokta kümesidir. "Bu şeklin genişliği ne?", "Yüksekliği ne?", "Tam olarak nerede başlıyor?" gibi soruların cevabını bu fonksiyonla alırız.
- 4 adet integer döndürür:
    1. X koordinatı
    2. Y koorddinatı
    3. En
    4. Boy

````python
#--AZ ÖNCEKİ KISIM--
for color_name,range in colors.items():
    mask = cv2.inRange(image,range[0],range[1])
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for contour in contours:
        M = cv2.moments(contour)
        if M["m00"] != 0:
            cx = int(M["m10"]/M["m10"]) 
            cy = int(M["m01"]/M["m00"])
        else:
            cx, cy = 0,0
        cv2.circle(outputImage, (cx,cy), 5, (0,0,0), -1)

        #--YENİ KISIM--    
        x,y,w,h = cv2.boundingRect(contour)
        cv2.rectangle(outputImage, (x, y), (x + w, y + h), (0, 255, 0), 2)

cv2.imshow("Kareli",outputImage)
````
