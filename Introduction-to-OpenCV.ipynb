```python
#OpenCV Kütüphanesinin projeye dahil edilmesi
import cv2
```

### Resim Okuma

```python
#Bilgisayardaki resimler imread() kütüphanesiyle okunur.
image = cv2.imread('blocks.jpg')
```

### Resmin Yüksekliğini ve Genişliğini Çıkartma

```python
#shape fonksiyonu bize resmin boyutlarını bir tuple olarak berir:
#Bize sırasıyla şu 3 bilgiyi döndürür: Yükseklik, Genişlik, Kanal Sayısı (BGR yani 3)
#Normalde biz "Genişlik x Yükseklik" şeklinde kullanırız (1920x1080) ancak BURADA SIRALAMA TAM YERSİDİR!
(h, w) = image.shape[:2]
#[:2] ifadesi "baştan başla ve 2. index'e kadar al (2 dahil değil)" demektir.


print("Height = {}, Width = {}".format(h, w)) #Yüksekliği ve genişliği yazdırma.
```

### Görseldeki Bir Pixel'in RGB Değerini Çıkartma

```python
#OpenCV RGB değil BGR kullanır yani sıralama RGB'nin tam tersidir.
(B, G, R) = image[100,100] #Pixel, resmin matrisinden rastgele seçildi.
#Index belirtilmediği için bütün renk değerleri getirildi.

print("R = {}, G = {}, B = {}\n".format(R, G, B))

#image[Pixel'in-Y-Koordinatı, Pixel'in-X-Koordinatı, İstediğin-Rengin-Index'i] -> Blue=0, Green=1, Red=2
B = image[100,100,0]
G = image[100,100,1]
R = image[100,100,2]

print("R = {}, G = {}, B = {}".format(R, G, B))
```

### Resmin Belli Bir Bölümünü Çıkartma (Region of Interest)

```python
roi = image[100:500 , 200:700] #image[ Y_Başlangıç : Y_Bitiş , X_Başlangıç : X_Bitiş ]

roiRGB = cv2.cvtColor(roi, cv2.COLOR_BGR2RGB)

cv2.imshow("ROI Image", roi) #imshow("Pencere ismi", gösteriecelResim) ile pencere açarak gösterir
cv2.waitKey(0)  #Belirtilen süre kadar giriş bekler (0 sonsuz bekleme)
cv2.destroyAllWindows() #Bütün penvereleri kapatır.
```

### Resmin Boyutlarını Değiştirme

```python
#shape() gibi fonksiyonlarda matris mantığı geçerliyken: (satır,sütun) yani (yükseklik,genişlik)
#resize(), line() gibi fonksiyonlardan klasik (genişlik x yükseklik) geçerlidir.

resizeImage = cv2.resize(image, (500,500))
cv2.imshow("Resized Image", resizeImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Yukarıdaki kodda manuel en-boy verdiğimiz için oran-orantı kayıyordu, kaymaması için:
ratio = 800 / w #oranın hesaplanması: istenen/gerçekteki
resizeImagewithRatio = cv2.resize(image, (800 , int(h * ratio))) #yukarıda belirttiğim gibi (genişlik x yükseklik)
#ayrıca çarpma sonucu rasyonel çıkabileceği için en yakın int'e yuvarladık

cv2.imshow("Orantili Olarak Degistirilmis Boyutlu Gorsel" , resizeImagewithRatio)
cv2.imshow("Orijinal Gorsel" , image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Görselin Üzerine Dikdörtgen Çizmek: rectangle()

```python
#Bu işlem orijinal resme yerleştiği için kopyasını almak yerinde olacaktır
outputImage = image.copy()
outputImage = cv2.rectangle(outputImage, (200, 100), (600, 200), (255, 0, 0), 2)
#   outputImage: Değişikliğin üzerine yapıldığı görsel
#   (200,100): Dikddörtgenin sol üst köşesinin (x,y) konuları
#   (600,200): Dikddörtgenin sağ alt köşesinin (x,y) konuları
#   (255,0,0): Dikdörtgenin BGR'ye göre rengi
#   2:         Dikdörtgenin kalınlığı 

cv2.imshow("Dikdortgen Cisilmis Resim",outputImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Görselin Üzerinde Yazı Gösterme

```python
#Bu işlemden önce de resmin kopyasını alman gerekir ancak üstte zaten aldığım için şu an almayacağım
outputImage = cv2.putText(outputImage, 'github.com/Mertcangndn', (50,100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)

cv2.imshow("Yazi Yazilmis Resim",outputImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
